```python
import matplotlib.pyplot as plt
X = data['amount']
Y = data['cardholder_age']
plt.scatter(x=X , y=Y, marker = '*', c = 'red', alpha= 0.2,linewidths=5,edgecolors='black')
plt.scatter(x=X , y=Y, marker = '*', c = 'blue', alpha= 0.2, linewidths=10)
plt.title("amount v/s cardholder_age")
plt.xlabel("amount")
plt.ylabel("cardholder_age")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(4,4))
plt.subplot(1,2,1)
plt.scatter(x=X,y=Y,marker = 'o',c='green',alpha=0.8,linewidths=2,edgecolors='pink')
plt.xlabel("amount")
plt.ylabel("cardholder_age")
plt.subplot(1,2,2)
plt.scatter(x=X,y=data['merchant_category'],marker='o',c='blue',alpha=0.2,linewidths=2,edgecolors='red')
plt.show()
```

```python
plt.figure(figsize=(4,4))
plt.xlabel("amount")
plt.ylabel("cardholder_a")
plt.scatter(x=X,y=Y,marker = 'o',c='green',alpha=0.8,linewidths=2,edgecolors='pink')
plt.scatter(x=X,y=data['merchant_category'],marker='o',c='red',alpha=0.2,linewidths=2,edgecolors='black')
plt.xlabel("amount")
plt.ylabel("cardholder_age")
plt.grid()
plt.show()

```
